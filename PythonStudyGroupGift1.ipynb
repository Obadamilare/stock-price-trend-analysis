# Stock Price Trend Analysis

```python
#Task1: Create the variables
company_name = "NovaTech Plc"
stock_symbol = "NVT"
opening_price = 152.50
market_price = 168.75 #current market price
num_shares = 350 #numbers of shares owned
investor_target = 170.00 #investor target price
expected_growth = 8
daily_growth = 10 #let actual daily growth be rep. by 'daily_growth'
market_open = True
```

```python
#Task2: Calculate the price increase
price_increase = market_price - opening_price
print("Price increase: #", price_increase)
```

```python
#Task3: Calculate the current investment value
investment_val = market_price * num_shares
print("Current Investment Value: #", investment_val)
```

```python
#Task4: Update the current market price
market_price += 3.25
updated_price = market_price #Updated market price
print("Updated market price: #", updated_price)
```

```python
#Task5: Check if the target market price has beeen reached
target_reached = updated_price >= investor_target
print("Target Reached:", target_reached)
```

```python
#Task6: Check if the stock performed above expectation
above_expectation = daily_growth >= expected_growth #performance above expectation
print("Performance Above Expectation:", above_expectation)
```

```python
#Task7: Determine if the investment looks strong
investment_strong= market_open and target_reached
print("Investment Strong:", investment_strong)
```

```python
#Task8: Determine if the investor attention is needed
review_required = not investment_strong or not above_expectation
print("Investor Review Required:", review_required) 
```

```python
#Task9: Stock performance report
print("===============================================================")
print("DAILY STOCK PERFORMANCE REPORT")
print("===============================================================")
print("Company's Name:", company_name)
print("Stock Symbol:", stock_symbol)
print("Opening Price: #", opening_price)
print("Updated Market Price: #", updated_price)
print("Price Increase Today: #", price_increase)
print("Number of Shares:", num_shares)
print("Current Investment Value: #", investment_val)
print("Investor Target Price: #", investor_target)
print("Expected Growth:", expected_growth, "%")
print("Actual Growth:", daily_growth, "%")
print("Target Price Achieved:", target_reached)
print("Performance Above Expectations:", above_expectation) 
print("Market Open:", market_open)
print("Investment Strong:", investment_strong)
print("Investor Review Required:", review_required) 
print("===============================================================")
```
